Basic Ideas of "Python для анализа данных"
==

## Course details:

* Working language is English.
* Lectures are via online course [Python как иностранный](https://online.hse.ru/local/showcase/?cid=3523). **NB!** We expect that students will come to classes prepared. It means that student have watched videos and read tasks before the class.
* In addition to the online course offline lectures will be held. During the lectures students will consolidate the covered material, discuss solutions of problems from class work, and have an opportunity to ask questions
* All grades, handouts and other information will be published on the website [https://sss.krasilnikov.spb.ru/](https://sss.krasilnikov.spb.ru/). 
    * To login to the website you need to enter your corporate email address (yourusername@edu.hse.ru). A code will be sent to your email. Use it as the password. The code is valid for 5 minutes. As soon as you register in our telegram bot **@hsedatabot** (see information below), the code to enter the site will be sent to the telegram bot.
* Every week you can have additional practice and answer small multiple choice questions in our telegram bot **@python_millionare** to recall past topics. 

After completing the course, you will have to pass an independent programming exam (basic level).  
The Independent Exam $≠$ "Python для анализа данных" Exam
 


## Grades

* We have
  * two modules
  * five graded activities:
     *  mid-term exam  (paper-based, open book)
     * homework 
     * classwork (graded seminar in the last week of the second module)
     * exam  (paper-based, open book)
     * bonuses (be active during seminars, ask questions, answer teachers' questions)
  * each week there is a seminar task with zero weight - you can solve the task, upload Python notebook with solution to the website, and get the grade but it does not affect your final grade.
  
$$\textrm{final grade} = \textbf{0.2} * \textrm{mid-term exam} + \textbf{0.25} * \textrm{homework} + \textbf{0.15} * \textrm{classwork} + \textbf{0.4} * \textrm{exam} + \textbf{0.1} * \textrm{bonuses} $$
